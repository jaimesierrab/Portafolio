```python
import pandas as pd
import matplotlib.pyplot as plt
# Carga de datos desde URLs
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
# Agrupamos todas las tiendas en un diccionario
tiendas = {"tienda1": tienda1, "tienda2": tienda2, "tienda3": tienda3, "tienda4": tienda4}

# Métricas a calcular
ingresos = {}
categorias_populares = {}
promedio_calificaciones = {}
producto_mas_vendido = {}
producto_menos_vendido = {}
envio_promedio = {}

for nombre, df in tiendas.items():
    ingresos[nombre] = df["Precio"].sum()
    categorias_populares[nombre] = df["Categoría del Producto"].value_counts().idxmax()
    promedio_calificaciones[nombre] = df["Calificación"].mean()
    producto_mas_vendido[nombre] = df["Producto"].value_counts().idxmax()
    producto_menos_vendido[nombre] = df["Producto"].value_counts().idxmin()
    envio_promedio[nombre] = df["Costo de envío"].mean()
# Mostrar tablas de tuplas por tienda
for nombre in tiendas.keys():
    print(f"\nDatos de {nombre}:")
    print(("Ingresos totales", ingresos[nombre]))
    print(("Categoría más popular", categorias_populares[nombre]))
    print(("Promedio calificaciones", promedio_calificaciones[nombre]))
    print(("Producto más vendido", producto_mas_vendido[nombre]))
    print(("Producto menos vendido", producto_menos_vendido[nombre]))
    print(("Costo envío promedio", envio_promedio[nombre]))
# Gráfico 1: Ingresos por tienda
plt.bar(ingresos.keys(), ingresos.values())
plt.title("Ingresos por tienda")
plt.xlabel("Tienda")
plt.ylabel("Ingresos")
plt.show()
# Gráfico 2: Calificación promedio
plt.bar(promedio_calificaciones.keys(), promedio_calificaciones.values())
plt.title("Promedio de calificaciones por tienda")
plt.xlabel("Tienda")
plt.ylabel("Calificación promedio")
plt.show()
# Gráfico 3: Costo de envío promedio
plt.plot(envio_promedio.keys(), envio_promedio.values(), marker='o')
plt.title("Costo de envío promedio por tienda")
plt.xlabel("Tienda")
plt.ylabel("Costo de envío")
plt.show()
# Recomendación final (puede ajustarse con más lógica si deseas)
tienda_menor_ingreso = min(ingresos, key=ingresos.get)
print(f"\n🔍 Recomendación:")
print(f"El Sr. João debería considerar vender la {tienda_menor_ingreso}, ya que presenta el menor desempeño en términos de ingresos.")
```
